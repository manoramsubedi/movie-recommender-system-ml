#### Data Preprocessing

```python
# cols to be consider
# genres, id, keywords, title, overview, cast, crew
movies_df = movies_df[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
movies_df.head()
```

#### Handeling missing values

```python
movies_df.isnull().sum()
```

```python
movies_df = movies_df.dropna()
```

```python
movies_df.shape
```

```python
movies_df.iloc[0].genres
```

#### Converting the data into appropiate format and extracting necessary data

```python
import ast

def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

#### For genre column

```python
convert(movies_df.iloc[0].genres)
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
```

```python
movies_df['genres']
```

#### For keywords column

```python
movies_df.iloc[0].keywords
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

```python
movies_df['keywords']
```

#### For cast column

```python
movies_df['cast'][0]
```

```python
# to take first 3 actor names
def convert_cast(obj):
    counter = 0
    l = []
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert_cast)
```

```python
movies_df['cast']
```

#### For crew column

```python
movies_df['crew'][0]
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
fetch_director(movies_df['crew'][0])
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
movies_df['crew']
```

#### For overview column

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x: x.split())
```

```python
movies_df.head()
```

```python
for col in movies_df.columns:
    #print(col)
    if col != 'id':
        if col != 'title':
            movies_df[col] = movies_df[col].apply(lambda x:[i.replace(" ","") for i in x])
movies_df.head(3)
```

```python
movies_df['tags'] = movies_df['genres'] + movies_df['keywords'] + movies_df['overview'] + movies_df['cast'] + movies_df['crew']
```

```python
movies_df.head(3)
```

```python
movies_df['tags']
```

```python
new_df = movies_df[['id', 'title', 'tags']]
new_df.head()
```

```python
# Changing tags column into string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
# into lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

#### Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y) # list into string
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

#### Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
vector
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

```python
# for i in cv.get_feature_names_out():
#     print(i)
features = cv.get_feature_names_out().tolist()
print(features[:100])
```

#### Using cosine distance to find similar movies

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
# Function to recommend movie based off given movie
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    print("Recommend Movies:")
    for movie in movies_list:
        print(new_df.iloc[movie[0]]['title'])
        
```

```python
recommend("Avatar")
```
